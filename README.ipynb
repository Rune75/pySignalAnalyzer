# pcmAnalyzer is a python class to run spectrum analysis on sampled data

#### Class usage
Create a test data vector with sampling specification.

```python
import sys
sys.path.append("test")
from pcmVector import pcmSignal


# create a coherent test signal
pcmSgl = pcmSignal(frequency=355e3,                 # signal frequency Hz the signal frequency will be adjusted for coherence
                    amplitude=0,                    # signal amplitude dBFS
                    sampling_rate=1e6,              # sampling rate S/s   
                    adcResolution=12,               # ADC resolution in bits
                    harmonic_levels=[-76, -80, -76, -80],   # harmonic levels in dBFS
                    nrSamples=8192                  # number of samples
                    )
```

Create a pcmAnalyzer object initialized with the sampled signal from the pcmSignal class and the accompanying sampling_rate and adcResolution. 

```python
from pcmAnalyzer import pcmAnalyzer  # import the analyzer class

# Initialise the pcmAnalyzer class with the test signal in pcmSgl.pcmVector
analysis = pcmAnalyzer(pcmSgl.pcmVector, pcmSgl.sampling_rate, pcmSgl.adcResolution)

# Plot the power spectrum and the scalar results
analysis.plotPowerSpectrum()
print('')
```

```python
analysis.printAll()
```

```python
# Use getters to get the results
print('SNR: ', analysis.getSNR())
print('THD: ', analysis.getTHD())
print('SFDR: ', analysis.getSFDR())
print('ENOB: ', analysis.getENOB())
print('SNR: ', analysis.getSNR())
print('SNDR_FS: ', analysis.getSNDR_FS())
print('SNDR_dB: ', analysis.getSNDR())
print('ENOB: ', analysis.getENOB())
print('ENOB_FS: ', analysis.getENOB_FS())
```
