# pcmAnalyzer is a python class to run spectrum analysis on sampled data
The class usage is shown below

```python
from pcmAnalyzer import pcmAnalyzer
from pcmVector import pcmSignal

# create a test signal
pcmSgl = pcmSignal(frequency=87e3, 
                    amplitude=0, 
                    sampling_rate=1e6, 
                    adcResolution=12,
                    harmonic_levels=[-90, -80, -76, -80]
                    )

pcmSgl.plotPCMVector()
pcmSgl.printSignalSpecs()
```

```python
# Initialise the pcmAnalyzer class with the test signal
analysis = pcmAnalyzer(pcmSgl.pcmVector, pcmSgl.sampling_rate, pcmSgl.adcResolution)

# Plot the power spectrum and the static results
analysis.plotPowerSpectrum()
```
