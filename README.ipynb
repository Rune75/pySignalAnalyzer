# pcmAnalyzer is a python class to run spectrum analysis on sampled data
Class usage:

First we need to create a tes data vector with sampling specification by initializing this pcmSignal class with our signal and sampling specification the pcmSignal class wil adjust the input frequency to achieve coherent sampling

```python
from pcmVector import pcmSignal     # import the simulated ADC and signal source

# create a coherent test signal
pcmSgl = pcmSignal(frequency=87e3,                  # signal frequency Hz
                    amplitude=0,                    # signal amplitude dBFS
                    sampling_rate=1e6,              # sampling rate S/s   
                    adcResolution=12,               # ADC resolution in bits
                    harmonic_levels=[-90, -80, -76, -80],   # harmonic levels in dBc
                    nrSamples=8192                  # number of samples
                    )
# print some signal specs
pcmSgl.printSignalSpecs()
```

Next we can initialize the pcmAnalyzer class with the sampled signal from the pcmSignal class and the accompanying sampling_rate and adcResolution. This is where we would input some real ADC data instead.

```python
from pcmAnalyzer import pcmAnalyzer  # import the analyzer class

# Initialise the pcmAnalyzer class with the test signal
analysis = pcmAnalyzer(pcmSgl.pcmVector, pcmSgl.sampling_rate, pcmSgl.adcResolution)

# Plot the power spectrum and the scalar results
analysis.plotPowerSpectrum()
```
